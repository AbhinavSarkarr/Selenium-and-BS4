```python
#scarping all the laptops name 

import requests
from bs4 import BeautifulSoup

url = 'https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops'

re = requests.get(url)

soup = BeautifulSoup(re.text, 'lxml')

laptops = soup.find_all('a', class_='title')
```

```python
for i in laptops:
    print(i.string.strip())
```

```python
#using regular expression 

acer = soup.find_all(string='Acer')  #exact word
```

```python
acer
```

```python
import re
acer = soup.find_all(string=re.compile('Acer')) #re.compile('Acer') is a regex that will match any string containing the word 'Acer'
print(acer)
```
