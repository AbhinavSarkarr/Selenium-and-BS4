```python
#lets start scraping the first page
 
import pandas as pd
url = 'https://www.flipkart.com/search?q=mobiles+under+50+k&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1'
headers = {
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
'Accept-Language': 'en-US,en;q=0.5',
'Accept-Encoding': 'gzip, deflate, br',
'Connection': 'keep-alive',
'Upgrade-Insecure-Requests': '1'
}
re = requests.get(url, headers=headers)
print(re.status_code)

soup = BeautifulSoup(re.text, 'lxml')

box = soup.find('div', class_='DOjaWF gdgoEp')

names = box.find_all('div', class_='KzDlHZ')
fomatted_names = []
for i in names:
    fomatted_names.append(i.text)

price = box.find_all('div', class_='Nx9bqj _4b5DiR')
formatted_price = []
for i in price:
    formatted_price.append(i.text)

desc = box.find_all('ul', class_='G4BRas')
formatted_desc = []
for i in desc:
    formatted_desc.append(i.text)

rating = box.find_all('div', class_='XQDdHH')
formatted_rating = []
for i in rating:
    formatted_rating.append(i.text)


df = pd.DataFrame({'Product Name': fomatted_names, 'Price': formatted_price, 'Description': formatted_desc, 'Rating': formatted_rating})

print(df)
```

```python
#Now let's perform this for multiple pages 
fomatted_names = []
formatted_price = []
formatted_desc = []
formatted_rating = []


for i in range(1,437):
    url = f'https://www.flipkart.com/search?q=mobiles+under+50+k&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}'
    headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1'
    }
    re = requests.get(url, headers=headers)

    soup = BeautifulSoup(re.text, 'lxml')
    box = soup.find('div', class_='DOjaWF gdgoEp')

    names = box.find_all('div', class_='KzDlHZ')
    for i in names:
        fomatted_names.append(i.text)

    price = box.find_all('div', class_='Nx9bqj _4b5DiR')
    for i in price:
        formatted_price.append(i.text)

    desc = box.find_all('ul', class_='G4BRas')
    for i in desc:
        formatted_desc.append(i.text)

    rating = box.find_all('div', class_='XQDdHH')
    for i in rating:
        formatted_rating.append(i.text)

print(len(fomatted_names), len(formatted_price), len(formatted_desc), len(formatted_rating ))
df = pd.DataFrame({'Product Name': fomatted_names, 'Price': formatted_price, 'Description': formatted_desc, 'Rating': formatted_rating})
df.to_csv('flipkart_products.csv', index=False, encoding='utf-8')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

formatted_names = []
formatted_price = []
formatted_desc = []
formatted_rating = []

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    'Accept-Language': 'en-US,en;q=0.5',
    'Accept-Encoding': 'gzip, deflate, br',
    'Connection': 'keep-alive',
    'Upgrade-Insecure-Requests': '1'
}

for i in range(1, 437):
    url = f'https://www.flipkart.com/search?q=mobiles+under+50+k&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}'
    re = requests.get(url, headers=headers)
    soup = BeautifulSoup(re.text, 'lxml')

    box = soup.find('div', class_='DOjaWF gdgoEp')

    if box is not None:
        names = box.find_all('div', class_='KzDlHZ')
        if names:
            for name in names:
                formatted_names.append(name.text)
        else:
            formatted_names.append("NA")
        
        price = box.find_all('div', class_='Nx9bqj _4b5DiR')
        if price:
            for pr in price:
                formatted_price.append(pr.text)
        else:
            formatted_price.append("NA")
        
        desc = box.find_all('ul', class_='G4BRas')
        if desc:
            for d in desc:
                formatted_desc.append(d.text)
        else:
            formatted_desc.append("NA")

        rating = box.find_all('div', class_='XQDdHH')
        if rating:
            for rate in rating:
                formatted_rating.append(rate.text)
        else:
            formatted_rating.append("NA")

    print(f"Page {i}: {len(formatted_names)} names, {len(formatted_price)} prices, {len(formatted_desc)} descriptions, {len(formatted_rating)} ratings")

# After the loop, check if all lists have the same length
max_length = max(len(formatted_names), len(formatted_price), len(formatted_desc), len(formatted_rating))

formatted_names.extend(["NA"] * (max_length - len(formatted_names)))
formatted_price.extend(["NA"] * (max_length - len(formatted_price)))
formatted_desc.extend(["NA"] * (max_length - len(formatted_desc)))
formatted_rating.extend(["NA"] * (max_length - len(formatted_rating)))

df = pd.DataFrame({'Product Name': formatted_names, 'Price': formatted_price, 'Description': formatted_desc, 'Rating': formatted_rating})
df.to_csv('flipkart_products.csv', index=False, encoding='utf-8')
```
